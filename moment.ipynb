```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Diabetes Classification.csv")
df.head()
```

```python
target_1 = df[df['Diagnosis'] == 1]
target_0 = df[df['Diagnosis'] == 0]
print("Data for Target 1:")
print(target_1.head())
print("\nData for Target 0:")
print(target_0.head())
```

```python
target_1=target_1.drop(columns="Diagnosis")
target_0=target_0.drop(columns="Diagnosis")
```

```python
target_1 = target_1.loc[:, ~target_1.columns.str.contains('^Unnamed')]
target_0 = target_0.loc[:, ~target_0.columns.str.contains('^Unnamed')]
```

```python
target_1=target_1.drop(columns="Gender")
target_0=target_0.drop(columns="Gender")
target_1.head()
```

```python
skewness_1=target_1.skew()
skewness_0=target_0.skew()
```

```python
print(skewness_1)
```

```python
print(skewness_0)
```

```python
mean_1= target_1.mean()
variance_1 = target_1.var()
skewness_1 = target_1.skew()
kurtosis_1 = target_1.kurtosis()

# Print the calculated moments
print("Mean:")
print(mean_1)
print("\nVariance:")
print(variance_1)
print("\nSkewness:")
print(skewness_1)
print("\nKurtosis:")
print(kurtosis_1)
```
