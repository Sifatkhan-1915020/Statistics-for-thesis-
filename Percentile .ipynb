```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Diabetes Classification.csv")
df.head()
```

```python

target_1 = df[df['Diagnosis'] == 1]
target_0 = df[df['Diagnosis'] == 0]
print("Data for Target 1:")
print(target_1.head())
print("\nData for Target 0:")
print(target_0.head())
```

```python
HDL_Target_1=target_1["HDL"]
percentile_50_1=np.percentile(HDL_Target_1,50)
HDL_Target_0=target_0["HDL"]
percentile_50_0=np.percentile(HDL_Target_0,50)

print("\n HDL Percentile for Diabetes Patietn ")
print(percentile_50_1)
print("\n HDL Percentile for Non Diabetes Patietn ")
print(percentile_50_0)
```
