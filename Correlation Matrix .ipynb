```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Diabetes Classification.csv")
df.head()
```

```python
target_1 = df[df['Diagnosis'] == 1]
target_0 = df[df['Diagnosis'] == 0]
print("Data for Target 1:")
print(target_1.head())
print("\nData for Target 0:")
print(target_0.head())
```

```python
target_1=target_1.drop(columns="Diagnosis")
target_0=target_0.drop(columns="Diagnosis")
```

```python
target_1 = target_1.loc[:, ~target_1.columns.str.contains('^Unnamed')]
target_0 = target_0.loc[:, ~target_0.columns.str.contains('^Unnamed')]
```

```python
target_1=target_1.drop(columns="Gender")
target_0=target_0.drop(columns="Gender")
target_1.head()
```

```python
corrMat_1=round(target_1.corr(),2)
corrMat_0=round(target_0.corr(),2)
```

```python
print(corrMat_1)
```

```python
print(corrMat_0)
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
axis_corr_1 = sns.heatmap(
corrMat_1,
vmin=-1, vmax=1, center=0,
cmap=sns.diverging_palette(50, 500, n=500),
square=True
)

plt.show()
```

```python
axis_corr_0=sns.heatmap(
    corrMat_0,vmin=-1, vmax=1, center=0,
cmap=sns.diverging_palette(50, 500, n=500),
square=True
)
```
